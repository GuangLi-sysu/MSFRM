We focus on four real hacking events\footnote{https://rekt.news/leaderboard/} in cryptocurrency finance, and the targets of the attacks including Ronin network, BNB bridge, Euler finance and Harmony bridge. To evaluate MS-FRM on real-world data, we collect the some of transactions generated by the attacker, and build the real-world datasets by selecting the transactions during the happened date of hack events. Details about these datasets:

| Datasets   | #nodes    | #edges    | Case Date |
|------------|-----------|-----------|-----------|
| RONIN      | 1,386,806 | 4,177,581 | Mar-2022  |
| BNB BRIDGE | 1,243,567 | 4,275,286 | Oct-2022  |
| HARMONY    | 340,725   | 1,067,761 | Jan-2023  |
| EULER      | 630,469   | 1,972,189 | Mar-2023  |

```python
import msfrm
import pandas as pd
```

Run MS-FRM on real-world datasets(BNB, RONIN, EULER, HARMONY).


k is the set number of rank.

```python
cases = ['bnb', 'ronin', 'euler', 'harmony']
k = 1000

for case in cases:
    tensor = pd.read_csv(f'../dataset_rea/{case}_data.csv', header=None)
    tensor.columns = ['from','to', 'amount']
    result = msfrm.msfrm(tensor, k)
    with open(f'./output/{case}_log.txt', 'a') as f:
        for weights, flow in result:
            f.write(f'{weights}, {flow}')
            f.write('\n')
```
